```python
params = ["Key Passes", "Touches in Box", "Shot-Creating Actions", "Dribbles Completed"]

low =  [0, 0, 0, 0]
high = [10, 10, 10, 10]

radar = Radar(params, low, high,
              round_int = [False] * len(params),
              num_rings = 4,
              ring_width = 1, center_circle_radius = 1)

player_a = [1, 2, 3, 4]
player_b = [2, 4, 8, 8]

fig, ax = radar.setup_axis()

rings_inner = radar.draw_circles(ax = ax, facecolor = "cornflowerblue", edgecolor = "k")
radar_output = radar.draw_radar(player_a, ax = ax,
                                kwargs_radar = {"facecolor": "coral", "edgecolor": "k"},
                                kwargs_rings = {"facecolor": "coral", "edgecolor": "coral"})

radar_poly, rings_outer, vertices = radar_output
range_labels = radar.draw_range_labels(ax = ax, fontsize = 15)
param_labels = radar.draw_param_labels(ax = ax, fontsize = 15)

plt.show()
```

```python
# parameter names of the statistics we want to show
params = ["Key Passes", "Touches in Box", "Shot-Creating Actions", "Dribbles Completed"]

low =  [0, 0, 0, 0]
high = [10, 10, 10, 10]

radar = Radar(params, low, high,
              round_int = [False] * len(params),
              num_rings = 4,
              ring_width = 1, center_circle_radius = 1)

player_a = [7, 6, 2, 1]
player_b = [2, 4, 8, 8]


fig, ax = radar.setup_axis(facecolor = "None")

rings_inner = radar.draw_circles(ax = ax, facecolor = 'cornflowerblue', edgecolor = 'k')
radar_output = radar.draw_radar_compare(player_a, player_b, ax = ax,
                                        kwargs_radar = {'facecolor': 'coral', "alpha": 0.5, "edgecolor": "k"},
                                        kwargs_compare = {"facecolor": "lime", "alpha": 0.5, "edgecolor": "k"})

radar_poly, radar_poly2, vertices1, vertices2 = radar_output
range_labels = radar.draw_range_labels(ax = ax, fontsize=15, color = "w")
param_labels = radar.draw_param_labels(ax = ax, fontsize=15, color = "w")

fig.set_facecolor('#121212')

plt.show()
```
