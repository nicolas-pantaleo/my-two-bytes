## Imports

# Working with Titles

```python
fig, ax = plt.subplots(figsize = (6, 4))

plt.scatter(tips.tip, tips.total_bill)

plt.title(label = "Formula: $x^2+\sqrt{x}$",
          y = 1.1,
          loc = "left",
          ha = "left",
          va = "top",
          fontsize = 20,
          color = "coral",
          style = "normal")

plt.title("Mini title",
          y = 1.05,
          loc = "right",
          fontsize = 12,
          ha = "right",
          va = "top",
          color = "grey",
          style = "italic")

plt.suptitle("Main Title\nSecond Line",
            fontsize = 25,
            y = 1.15,
            ha = "center",
            va = "center")

# Da salvare
#plt.subplots_adjust(top=0.7)
plt.show()
```

# Working with Ticks & Axis

```python
fig, ax = plt.subplots(figsize = (6, 4))

plt.bar(np.arange(5), [3, 12, 5, 18, 45], color=(0.2, 0.4, 0.6, 0.6))
 
plt.xlabel(xlabel = "Asse X",
           fontweight = "bold",
           color = "orange",
           fontsize = "17",
           ha = "center")

plt.tick_params(axis = "x",
                colors = "red",
                direction = "out",
                length = 10,
                width = 3)

plt.tick_params(axis = "y",
                colors = "blue",
                direction = "in",
                length = 10,
                width = 2)

plt.xticks(ticks = np.arange(5),
           labels = list("ABCDE"),
           color = "coral",
           rotation = 45,
           fontweight = "bold",
           fontsize = 12,
           ha = "left")

plt.xlim(-1, 6)
plt.ylim(-0.5, 50)

plt.tick_params(top = True, right = True)

plt.show()
```

# This is really random

```python
df = pd.DataFrame({"A": range(100),
                   "B": np.random.randn(100) * 15 + range(100)})

fig, ax = plt.subplots(figsize = (8, 4))

plt.scatter("A", "B", data = df)

rnd_obs = df.sample(n = 1)
rnd_x, rnd_y = rnd_obs.A.values[0], rnd_obs.B.values[0]

plt.annotate(s = "$x^2 - \infty$",
             xy = (rnd_x, rnd_y),
             xytext = (0, 100),
             fontsize = 20,
             arrowprops = {"facecolor": "k", "shrink": 0.05})

rect_width = np.percentile(a = df.A, q = 75, interpolation = "nearest") \
             - np.percentile(a = df.A, q = 25, interpolation = "nearest")

rect_height = np.percentile(a = df.B, q = 75, interpolation = "nearest") \
             - np.percentile(a = df.B, q = 25, interpolation = "nearest")

np.percentile(a = df.A, q = 75, interpolation = "nearest")

ax.add_patch(plt_patches.Rectangle(xy = (
    np.percentile(a = df.A, q = 25, interpolation = "nearest"),
    np.percentile(a = df.B, q = 25, interpolation = "nearest")
    ),
    width = rect_width,
    height = rect_height,
    alpha = 0.3, facecolor = "coral", edgecolor = "k",
    linewidth = 4, linestyle = "solid"))


plt.axvline(x = np.mean(df.A))
plt.axhline(y = np.median(df.B))
plt.show()
```

# Two Columns Subplot

```python
df = pd.DataFrame({"col1": range(100),
                   "col2": np.random.randn(100) * 15 + range(100)})

fig, ax = plt.subplots(figsize = (8, 4))

plt.subplot(121)
plt.plot("col1", "col2", data = df)

plt.subplot(122)
plt.plot("col1", "col2", data = df, color = "coral")

plt.show()
```

# Two Rows Subplot \#2

```python
df = pd.DataFrame({"col1": range(100),
                   "col2": np.random.randn(100) * 15 + range(100)})

fig, axes = plt.subplots(nrows = 2, ncols = 1,
                         sharex = True, sharey = True,
                         figsize = (8, 4))

axes[0].plot("col1", "col2", data = df)
axes[1].plot("col1", "col2", data = df, color = "coral")

axes[0].title.set_text("Asse x in comune")

plt.show()
```

# Matrix (2x2)

```python
df = pd.DataFrame({"col1": range(100),
                   "col2": np.random.randn(100) * 15 + range(100)})

fig, axes = plt.subplots(nrows = 2, ncols = 2,
                         figsize = (8, 4))

axes[0, 0].plot("col1", "col2", data = df)
axes[0, 1].plot("col1", "col2", data = df, color = "coral")
axes[1, 0].plot("col1", "col2", data = df)
axes[1, 1].plot("col1", "col2", data = df, color = "coral")

axes[0, 0].title.set_text("Asse x in comune")

plt.suptitle("Titolo di tutto", y = 1.05)

plt.show()
```
