## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as plt_patches
import matplotlib.patheffects as plt_eff
import seaborn as sns
```

# Basic Spiderplot

```python
df = pd.DataFrame({
    "group": list("ABCD"),
    "var1": [38, 1.5, 30, 4],
    "var2": [29, 10, 9, 34],
    "var3": [8, 39, 23, 24],
    "var4": [7, 31, 33, 14],
    "var5": [28, 15, 32, 14]})

categories = len(list(df)) - 1

obs = df.iloc[0].drop("group").values.flatten().tolist()
obs.append(obs[0])

angles = [n / float(categories) * 2 * np.pi for n in range(categories)]
angles.append(angles[0])
```

```python
fig, ax = plt.subplots(figsize = (8, 6))

plt.subplot(111, polar = True)
plt.xticks(angles[:-1], list(df)[1:], color = "grey", size = 12)
plt.gca().set_rlabel_position(0)
plt.gca().tick_params(axis = "x", pad = 30)
plt.yticks([10, 20, 30, 40, 50], ["10", "20", "30", "40", "50"],
           color = "grey", size = 12)
plt.ylim(0, 60)
plt.plot(angles, obs, linewidth = 1, linestyle = "solid")
plt.fill(angles, obs, color = "coral", alpha = 0.1)

plt.show()
```

# Spider Chart Playground

```python
fig, ax = plt.subplots(figsize = (8, 6))

plt.subplot(111, polar = True)
plt.bar(x = np.pi / 8,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "coral")

plt.bar(x = (np.pi/4 + np.pi/2)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "palegreen")

plt.bar(x = (3*np.pi/4 + np.pi/2)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "teal")

plt.bar(x = (np.pi + 3*np.pi/4)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "cornflowerblue")

plt.bar(x = (np.pi + 5*np.pi/4)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "gold")

plt.bar(x = (3*np.pi/2 + 5*np.pi/4)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "purple")

plt.bar(x = (3*np.pi/2 + 7*np.pi/4)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "grey")

plt.bar(x = (7*np.pi/4 + 2*np.pi)/2,
        height = 10, width = np.pi / 4,
        bottom = 0, color = "skyblue")

plt.show()
```
