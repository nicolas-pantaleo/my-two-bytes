# Matplotlib & Seaborn - Histograms

---

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as plt_patches
import matplotlib.patheffects as plt_eff
import seaborn as sns
```

---

## Datasets

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

---

---

# Basic Histograms

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.histplot(data = iris, x = "sepal_length",
             kde = True, bins = 20,
             color = "cornflowerblue")

sns.histplot(data = iris, x = "sepal_width",
             kde = True, bins = 20,
             color = "coral")

plt.show()
```

---

# Histogram with *Rug*

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.histplot(data = tips, x = "total_bill",
             kde = True, bins = 20)

sns.rugplot(data = tips, x = "total_bill",
            height = 0.1, alpha = 0.4, linewidth = 3)

plt.show()
```

---

# Horizontal Histogram

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.histplot(data = tips,
             y = "total_bill",
             bins = 20,
             kde = True)

plt.show()
```

---

# Histogram with Boxplot on top

```python
fig, (ax_box, ax_hist) = plt.subplots(2, sharex = True, figsize = (8, 4), gridspec_kw = {"height_ratios": (0.15, 0.85)})

sns.boxplot(x = tips.total_bill, ax = ax_box)
sns.histplot(x = tips.total_bill, ax = ax_hist)

ax_box.set(xlabel = None, xticks = [], yticks = [])
plt.show()
```

---

# Histogram Grid

```python
fig, axes = plt.subplots(2, 2, figsize = (8, 8))

sns.histplot(data = iris, x = "sepal_length", kde = True, color = "skyblue", ax = axes[0, 0])
sns.histplot(data = iris, x = "sepal_width", kde = True, color = "olive", ax = axes[0, 1])
sns.histplot(data = iris, x = "petal_length", kde = True, color = "gold", ax = axes[1, 0])
sns.histplot(data = iris, x = "petal_width", kde = True, color = "teal", ax = axes[1, 1])

axes[0, 1].set(ylabel = "")
axes[1, 1].set(ylabel = "")
plt.show()
```

---

# KDEPlot with different *bw_method*

```python
fig, axes = plt.subplots(2, 2, figsize = (10, 6))

sns.kdeplot(x = tips.total_bill,
            shade = True, bw_method = "scott",
            color = "olive", ax = axes[0, 0])

sns.kdeplot(x = tips.total_bill,
            shade = True, bw_method = 0.2,
            color = "teal", ax = axes[0, 1])

sns.kdeplot(x = tips.total_bill,
            shade = True, bw_method = 0.1,
            color = "coral", ax = axes[1, 0])

sns.kdeplot(x = tips.total_bill,
            shade = True, bw_method = 0.05,
            color = "cornflowerblue", ax = axes[1, 1])

plt.show()
```

---

# Double KDEPlot

```python
fig, ax = plt.subplots(figsize = (8, 4))

sns.kdeplot(data = iris,
            x = "sepal_width",
            shade = True, color = "coral")

sns.kdeplot(data = iris,
            x = "sepal_length",
            shade = True, color = "cornflowerblue")

plt.show()
```
