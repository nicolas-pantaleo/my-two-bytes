# Matplotlib & Seaborn - Pieplots

---

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as plt_patches
import matplotlib.patheffects as plt_eff
import seaborn as sns
```

---

## Datasets

```python
tips = sns.load_dataset("tips")
```

---

---

# Basic Pieplot

```python
fig, ax = plt.subplots(figsize = (6, 4))

plt.pie(tips.day.value_counts().values,
        radius = 1.5,
        wedgeprops = {"edgecolor":"k",
                      "linewidth": 1,
                      "linestyle": "-",
                      "antialiased": True})

plt.show()
```

---

# Basic Donutplot

```python
fig, ax = plt.subplots(figsize = (6, 4))

plt.pie(tips.day.value_counts().values,
        labels = tips.day.value_counts().index,
        radius = 1.5,
        colors = ["coral", "cornflowerblue", "palegreen", "gold"],
        labeldistance = 1.1,
        textprops = {"fontsize": 14, "color": "k"},
        wedgeprops = {"edgecolor":"k",
                      "linewidth": 1,
                      "linestyle": "-",
                      "antialiased": True})

circle = plt.Circle((0, 0), radius = 1,
                    facecolor = "white",
                    edgecolor = "k")

fig.gca().add_artist(circle)

plt.show()
```

---

# Borderless Donutplot 

```python
fig, ax = plt.subplots(figsize = (6, 4))

plt.pie(tips.day.value_counts().values,
        labels = tips.day.value_counts().index,
        radius = 1.5,
        colors = ["coral", "cornflowerblue", "palegreen", "gold"],
        labeldistance = 1.1,
        textprops = {"fontsize": 14, "color": "k"},
        wedgeprops = {"edgecolor":"white",
                      "linewidth": 7,
                      "linestyle": "-",
                      "antialiased": True})

circle = plt.Circle((0, 0), radius = 1,
                    facecolor = "white",
                    edgecolor = "white")

fig.gca().add_artist(circle)

plt.show()
```
